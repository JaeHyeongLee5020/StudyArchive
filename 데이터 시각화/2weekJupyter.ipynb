```python
import csv
import matplotlib.pyplot as plt

# CSV 파일 열기
with open('suwon.csv', encoding='cp949') as f:
    data = csv.reader(f, delimiter=',')
    header = next(data)
    
    # 결과를 저장할 리스트 초기화
    result = []
    
    # 데이터 처리 루프
    for row in data:
        # 마지막 열이 실수로 변환 가능한지 확인하고 처리
        if row[-1].replace(',', '', 1).isdigit():
            result.append(float(row[-1]))
    
    # 처리된 데이터의 개수 출력
    print(len(result))
    
    # 그래프 그리기
    plt.plot(result, 'r')
    plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
f = open('suwon.csv',encoding = 'cp949')
data = csv.reader(f, delimiter = ',')
header = next(data)
result_high = []
result_low = []
for row in data:
    if row[-1] != '':
           if row[0].split('-')[0] >= '\t1990': 
                if row[0].split('-')[1] == '06' and row[0].split('-')[2] == '15':
                    result_high.append(float(row[-1]))
                    result_low.append(float(row[-2]))
                    
plt.figure(figsize=(10,4))
plt.plot(result_high,'r',ls='-', label='circle')
plt.plot(result_low,'b^',ls=':',label='dotted')
plt.legend()
plt.show()
```
